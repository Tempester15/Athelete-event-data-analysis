- function __encode_sex()__ encodes the gender in the manner
    1. male = 1
    2. female = 0

```python
def encode_sex(val):
    if val == 'M':
        return 1
    elif val == 'F':
        return 0
    else:
        return np.nan
data.Sex = data.Sex.apply(encode_sex)
data
```

 - since the data in the columns __'Age'__, __'Height'__, __'Weight'__ is of different scale, therefore normalizing the data with __z-score__.
 - z-score = (x - $\mu$ ) / $\sigma$

```python
cols = ['Age', 'Height', 'Weight']
for col in cols:
    mean_val = data1[col].mean()
    std_val = data1[col].std()
    norm_val = (data1[col] - mean_val) / std_val
    data1[col] = norm_val
    

    
sns.boxplot(data = data1[cols])
```

 - Hence the all the numeric data came in same range

```python
# number of unique values in each categorical column
def print_unique(col):
    print(f'unqiue values in {col} : {data1[col].unique().shape}')
    
cols = ['Team', 'NOC', 'Games', 'Year', 'Season', 'City', 'Sport', 'Event', 'Medal']

for col in cols:
    print_unique(col)
```

```python
data1.drop(columns = ['Games'], inplace = True)
data1
```

 - function __encode_season()__ encodes the season column in the following manner : 
     1. Winter = 1
     2. Summer = 0

```python
def encode_season(val):
    if val == 'Summer':
        return 0
    elif val == 'Winter':
        return 1
    else:
        return np.nan
    

data1.Season = data1.Season.apply(encode_season)
data1
```

```python
data2 = data1.reset_index()
data2.drop(columns = 'ID', inplace = True)
data2
```

Encoding the __Medal__ column as:
- __Gold__ = 1
- __Silver__ = 2
- __Bronze__ = 3
- __no medal or NaN values__ = 0

```python
arr = np.zeros(len(data2))

gold = data2[data2.Medal == 'Gold'].index.tolist()
arr[gold] = 1

silver = data2[data2.Medal == 'Silver'].index.tolist()
arr[silver] = 2

bronze = data2[data2.Medal == 'Bronze'].index.tolist()
arr[bronze] = 3

arr = arr.astype('int')
data2['Medal'] = arr
data2
```

## Gold medal analysis

```python
def medal_analysis(medal_code):
    teams = data2.Team.unique().tolist()
    teams.sort()
    dic = {}
    for team in teams:
        x = data2[data2.Team == team]

        y = x[x.Medal == medal_code].shape[0]
        if y != 0:
            dic.update({f'{team}' : y})
        else:
            pass
    df = pd.DataFrame(columns = ['team', 'medals'])
    df['team'] = dic.keys()
    df['medals'] = dic.values()
    df = df.sort_values(by = 'medals', ascending = False)
    index = np.where(df['medals'] > 10)[0].tolist()
    plt.figure(figsize = (15, 15))
    ax = sns.barplot(x = df.iloc[index, 1], y = df.iloc[index, 0])
    for i in ax.containers:
        ax.bar_label(i)
    medals = {'Gold' : 1, 'Silver' : 2, 'Bronze' : 3}
    plt.ylabel('Team names')
    medal_name = list(medals.keys())[list(medals.values()).index(medal_code)] 
    plt.xlabel(f'No. of {medal_name} medals')
    plt.title(f'Teams that have won more than 10 {medal_name} medals')
    plt.show()
```

```python
medal_analysis(1)
```

## Silver medal analysis

```python
medal_analysis(2)
```

## Bronze medal analysis

```python
medal_analysis(3)
```

## Year wise analysis of medals

```python
data3 = data2[data2.Medal != 0].copy()

plt.figure(figsize = (20, 20))
ax = sns.countplot(data = data3, 
                   y = 'Year', 
                   hue = 'Medal', 
                   palette = ['#FFD700', '#708090', 'brown'], 
                   width = 0.7,
                  )

for i in ax.containers:
    ax.bar_label(i, size = 10,rotation = -15)
    
plt.xlabel('year', fontsize = 14)
plt.xticks(size = 12)
plt.ylabel('number of medals won in respected year', fontsize = 14)
plt.yticks(size = 12)
plt.legend(['Gold', 'gray', 'Bronze'], fontsize = 12)
```

- the above countplot tells us about year wise medal aquisition 
- Hence, we can see that 
    - the maximum number of __gold__ and __silver__ medals were obtained in the year __1984__.
    - the maximum number of __bronze__ medals were obtained in the year __1992__.

## Sports-wise medal analysis

- In the above graph you can see the sports vs number of medals.
- the above graph give the information about the total number of medals won in each sport.
- Hence we can see that the maximum numbers of __gold, silver__ and __bronze__ medals were obtained in the sport __Rowing__.
